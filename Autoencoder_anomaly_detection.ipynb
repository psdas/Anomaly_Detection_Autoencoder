```
from keras.layers import Input, Dense, Lambda, Dropout
from keras.models import Model,save_model,load_model
from keras.objectives import binary_crossentropy
from keras.callbacks import LearningRateScheduler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras.backend as K
import tensorflow as tf
from sklearn import preprocessing
import keras.optimizers
from tensorflow import set_random_seed
from numpy.random import seed
from sklearn.model_selection import train_test_split
import seaborn as sns

#-------Data Preproccessing-------#
def preprocess(df):

    #index=df[:,0:2]
    preprocessed_data=df
    scaler = preprocessing.MinMaxScaler()
    preprocessed_data = scaler.fit_transform(preprocessed_data)
    bol=np.equal.reduce(preprocessed_data)
    bol=np.invert(bol)
    #print(bol.shape)
    preprocessed_data=preprocessed_data[:,bol]
    mask = np.all(np.isnan(preprocessed_data), axis=1)
    mask=np.invert(mask)
    preprocessed_data=preprocessed_data[mask]
    #print(np.sum(bol))
    #index=index[mask]
    return preprocessed_data


#----------Training----------#
def model_create(data,m=50,n_z=6,n_epoch=100,split=0.1):
    input_shape=data.shape[1]
    i_s1=round(input_shape*0.75)
    i_s2=round(input_shape*0.5)
    i_s3=round(input_shape*0.25)

    #dat=data
    #train_data, test_data = train_test_split(data, test_size=0.3, random_state=0)
    
    train_data=data
    test_data=data
    #data_test=data[7500:]
    #data=data[0:7500]


    seed(10)
    set_random_seed(10)

    # Q(z|X) -- encoder
    inputs = Input(shape=(input_shape,))
    h_q1 = Dense(i_s1, activation='relu')(inputs)
    h_q2 = Dropout(0.2)(h_q1)
    h_q3 = Dense(i_s2, activation='relu')(h_q2)
    h_q4 = Dropout(0.2)(h_q3)
    h_q5 = Dense(i_s3, activation='relu')(h_q4)
    h_q6 = h_q2 = Dropout(0.2)(h_q5)
    h_q = Dense(n_z, activation='relu')(h_q6)
    mu = Dense(n_z, activation='sigmoid')(h_q)
    log_sigma = Dense(n_z, activation='sigmoid')(h_q)

    def sample_z(args):
        mu, log_sigma = args
        eps = K.random_normal(shape=(1, n_z), mean=0., stddev=1.)
        return mu + K.exp(log_sigma / 2) * eps

    # Sample z ~ Q(z|X)
    z = Lambda(sample_z)([mu, log_sigma])

    decoder_hidden = Dense(i_s3, activation='relu')
    d_q = Dropout(0.2)
    d_q1 = Dense(i_s2, activation='relu')
    d_q2 = Dropout(0.2)
    d_q3 = Dense(i_s1, activation='relu')
    d_q4 = Dropout(0.2)
    decoder_out = Dense(input_shape, activation='sigmoid')

    h_p = decoder_hidden(z)
    h_p1 = d_q(h_p)
    h_p2 = d_q1(h_p1)
    h_p3 = d_q2(h_p2)
    h_p4 = d_q3(h_p3)
    h_p5 = d_q4(h_p4)
    outputs = decoder_out(h_p3)

    def vae_loss(y_true, y_pred):
        recon = K.sum(K.binary_crossentropy(y_true, y_pred), axis=1)
        kl = 0.5 * K.sum(K.exp(log_sigma) + K.square(mu) - 1. - log_sigma, axis=1)
        return (recon + kl)

    vae = Model(inputs, outputs)
    encoder = Model(inputs, mu)
    d_in = Input(shape=(n_z,))
    d_h = decoder_hidden(d_in)
    d_h1 = d_q(d_h)
    d_h2 = d_q1(d_h1)
    d_h3 = d_q2(d_h2)
    d_out = decoder_out(d_h3)
    decoder = Model(d_in, d_out)
    opt = keras.optimizers.Adam(lr=0.001)
    vae.compile(optimizer=opt, loss=vae_loss)
    history = vae.fit(train_data, train_data, batch_size=m, epochs=n_epoch,verbose=1)
    #history = vae.fit(data, data, batch_size=m, epochs=n_epoch, validation_data=(data_test, data_test), verbose=1)

    plt.plot(history.history['loss'], 'b', label='Training loss')
    #plt.plot(history.history['val_loss'], 'r', label='Validation loss')
    plt.legend(loc='upper right')
    plt.xlabel('Epochs')
    plt.ylabel('Loss, [mse]')
    plt.ylim([0, 400])
    plt.show()
    pred = vae.predict(data,m)
    
    return pred,test_data



data=np.asarray(df)
data=preprocess(data)
print(data.shape)
pred,test_data = model_create(data,m=100,n_z=6,n_epoch=100,split=0.0)
def vae_loss(y_true, y_pred):
     #""" Calculate loss = reconstruction loss + KL loss for each data in minibatch """
     # E[log P(X|z)]
    y_true=tf.convert_to_tensor(y_true,preferred_dtype=tf.float64)
    y_pred = tf.convert_to_tensor(y_pred,preferred_dtype=tf.float64)

    recon = K.sum(K.binary_crossentropy(y_true, y_pred), axis=1)
    # D_KL(Q(z|X) || P(z|X)); calculate in closed form as both dist. are Gaussian
    #kl = 0.5 * K.sum(K.exp(log_sigma) + K.square(mu) - 1. - log_sigma, axis=1)
    return recon
loss=vae_loss(test_data, pred)


```
